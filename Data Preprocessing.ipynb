```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\amitthakur3\Downloads\Machine-Learning-with-Python-and-Scikit-master\Machine-Learning-with-Python-and-Scikit-master\Datasets\Mall_Customers.csv")
```

```python
## Creation of Dependent and Independent Variable
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
print(x)
```

```python
x[:,1:2]
```

```python
print(y)
```

### Missing Data Treatment

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer.fit(x[:,2:4])
x[:,2:4] = imputer.transform(x[:,2:4])
```

```python
print(x)
```

### Categorical data --> one hot encoding

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers =[('encoder', OneHotEncoder(), [1])] ,remainder = 'passthrough')
x = np.array(ct.fit_transform(x))
print(x)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x, y, test_size = 0.2, random_state=1)
```

```python
print(x_train)
```

```python
print(x_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[:, 3:] =  sc.fit_transform(x_train[:, 3:])
x_test[:, 3:] =  sc.transform(x_test[:, 3:])
```
